```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
%matplotlib inline
```

```python
iris_df = datasets.load_iris()
```

```python
list(iris_df.target_names)
```

```python
X = iris_df.data
Y = iris_df.target

X_reduced = decomposition.PCA(n_components=3).fit_transform(X)

fig = plt.figure(figsize=(10, 8))
ax = Axes3D(fig, auto_add_to_figure=False, elev=-150, azim=120)
fig.add_axes(ax)

ax.scatter3D(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=Y, cmap = plt.cm.rainbow)

ax.set_title("First 3 transformed dimensions")
ax.set_xlabel("1st eigenvector")
ax.set_ylabel("2nd eigenvector")
ax.set_zlabel("3rd eigenvector")
```
