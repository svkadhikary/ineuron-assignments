```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm',
'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
```

```python
df
```

```python
#1>
for i, value in enumerate(df['FlightNumber']):
    if pd.isna(value):
        df.loc[i, 'FlightNumber'] = df.loc[i-1, 'FlightNumber'] + 10
df['FlightNumber'] = df['FlightNumber'].astype(int)

print(df)
print(df['FlightNumber'].dtype)
```

```python
#2>
df_temp = pd.DataFrame(columns=['From', 'To'])
df_temp['From'] = [item.split('_')[0] for item in df['From_To']]
df_temp['To'] = [item.split('_')[1] for item in df['From_To']]
df_temp
```

```python
#3>
df_temp['From'] = df_temp['From'].str.capitalize()
df_temp['To'] = df_temp['To'].str.capitalize()

df_temp
```

```python
#4>
df = df.drop('From_To', axis=1)
df = pd.concat([df, df_temp], axis=1)
df
```

```python
#5>
#program to determine max number of columns for delays
delays = {}
for li in df['RecentDelays']:
    for delay in li:
        if 'delay_' + str(li.index(delay) + 1) not in list(delays.keys()):
            delays['delay_' + str(li.index(delay) + 1)] = []
delays_df = pd.DataFrame(delays)
delays_df
```

```python
#helper function to return list of delays according to the indices in the RecentDelays column
def delays_(li, idx):
    final = []
    for delay in li:
        try:
            final.append(delay[idx])
        except IndexError as e:
            final.append(np.nan)
    return final
```

```python
#assigning delays index-wise in the delays dataframe
delays_df['delay_1'] = delays_(df['RecentDelays'], 0)
delays_df['delay_2'] = delays_(df['RecentDelays'], 1)
delays_df['delay_3'] = delays_(df['RecentDelays'], 2)
delays_df
```

```python
#Replacing RecentDelays column with seperate individual delays
df = pd.concat([df, delays_df], axis=1)
df = df.drop('RecentDelays', axis=1)
df
```
